# Bit Manipulation
Play around with bits

```python
#depedencies
import time
```

Function

```python
def count( value ):
    count = 0
    while value:
        value &= value - 1
        count += 1
    return count
```

Running

```python
start = time.time()
print( count(0b1111) )
print("----- %s seconds -----" % (time.time() - start))
```

Check if Power of 2

```python
def isPowerof2(num):
    return (num and not (num & (num - 1)))
```

```python
print( isPowerof2(10) )
print( isPowerof2(256) )
```

Bitwise Shift

```python
print( 8 << 1 )
print( 8 >> 1 )
```

```python
def isSet(N, i):
    if (N & ( 1 << i )):
        return True
    return False
```

```python
print( isSet(4, 2) ) # 4 = 100
```

```python
print( isSet(6, 0) ) # 6 = 110
```

Set a bit


```python
def set(N, i):
    return (N | (1 << i))
```

```python
print ( set(5, 1)) # 5 = 101, will become 111
```
